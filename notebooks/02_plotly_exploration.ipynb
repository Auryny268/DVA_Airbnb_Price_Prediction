Quick demo that Plotly can produce interactive map visualizations for this project.
Loads the preprocessed enriched dataset and renders a density heatmap of listing prices.

```python
import pandas as pd
import plotly.express as px
from pathlib import Path

PROCESSED = Path('../data/processed')
REPORTS = Path('../reports')
REPORTS.mkdir(exist_ok=True)

enriched = pd.read_csv(PROCESSED / 'listings_nov2025_enriched.csv')
print(f'Loaded {len(enriched):,} listings')
print(enriched[['latitude', 'longitude', 'price_numeric', 'neighbourhood_group_cleansed']].head())
```

```python
fig = px.density_mapbox(
    enriched,
    lat='latitude',
    lon='longitude',
    z='price_numeric',
    radius=14,
    center={'lat': 40.7128, 'lon': -74.0060},
    zoom=10,
    mapbox_style='open-street-map',
    title='NYC Airbnb Nightly Price Density — Nov 2025 (short-term rentals only)',
    color_continuous_scale='Viridis',
    range_color=[0, enriched['price_numeric'].quantile(0.95)],
    labels={'price_numeric': 'Nightly price ($)'},
)
fig.update_layout(height=650, margin={'r': 0, 't': 40, 'l': 0, 'b': 0})

out = REPORTS / 'price_heatmap.html'
fig.write_html(str(out))
print(f'Saved interactive heatmap → {out}')
fig.show()
```
